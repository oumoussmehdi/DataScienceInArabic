### subplots - رسوم متعددة ضمن نفس الشكل

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * (1/x1)

x2 = np.linspace(0.0, 2.0)
y2 = np.sin(2 * np.pi * x2)

#plt.subplot(2, 1, 1)
plt.subplot(211)
plt.plot(x1, y1, 'o-')
plt.title('two subplots')
plt.ylabel('f(t)')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('f(t)')


plt.show()
```

## <center style='color:blue'> pie Chart - مخطط دائري </center>


```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
snacks = 'pizza', 'hotdog', 'shawerma', 'tacos'
lovers = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()

ax1.pie(lovers, 
        labels=snacks,
        explode=explode, 
        autopct='%1.1f%%',
        shadow=True, 
        startangle=180
       )

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## <center style='color:blue'> Scatter plot - المخطط المبعثر </center>


```python
x =np.linspace(-1, 1, 50)
y = np.random.randn(50)
 
plt.scatter(x,y)
```

```python
x =np.linspace(-1, 1, 50)
y = np.random.randn(50)
c = np.random.randn(50)
s = y * 200

fig, ax = plt.subplots()
ax.scatter(x,y,
           c= c, 
           s= s, 
           alpha=0.5)


ax.set_xlabel(r'x', fontsize=15)
ax.set_ylabel(r'y', fontsize=15)
ax.set_title('scatter plot with volume and colors')

ax.grid(True)
fig.tight_layout()

plt.show()
```
