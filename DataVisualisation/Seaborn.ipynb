Seaborn library is built on top of the Matplotlib library, therefore you can easily combine the two.

it does introduces additional plot types and it makes the Matplotlib plots look lot better

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np 
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns

sns.set(style="ticks", 
        color_codes=True)
```

### Fifa19 dataset

```python
fifa = pd.read_csv('..\\data\\fifa19.csv',index_col=0)
print(fifa.shape)
fifa.head()
```

```python
fifa.columns
```

```python
fifa.dtypes
```

### <center style='color:blue'> Scattter plot </center>

```python
sns.lmplot(x='Strength', 
           y='Stamina', 
           data=fifa,
          fit_reg=False,
          )
```

### <center style='color:blue'> line chart </center>

```python
sns.lineplot(x='Vision', 
             y='Penalties', 
             data=fifa)
```

```python
data =  [1,5,3,2,6] * 20
time = range(100)

sns.tsplot(data, time)
```

### <center style='color:blue'> Bar Chart </center>

```python
g = sns.barplot(x="Name", 
            y="Age", 
            data=fifa[:10],
           )

g.set_xticklabels(labels=fifa['Name'])
```

```python
g = sns.barplot(x="Name", 
            y="Age", 
            data=fifa[:10],
           )

g.set_xticklabels(labels=fifa['Name'][:10],rotation=90)
```

```python
g = sns.factorplot("Nationality", 
                   data=fifa, 
                   kind="count", 
                   size=5, aspect=4,
                   color="b")
g.set_xticklabels(rotation=90)
```

```python
# plt.figure(figsize=(45,10))

g = sns.factorplot("Nationality", 
                   data=fifa[:100], 
                   aspect=1.5, 
                   kind="count", 
                   color="b")
g.set_xticklabels(rotation=90)
```

### Tips Dataset

```python
tips = sns.load_dataset("tips")
tips.head()
```

### <center style='color:blue'> Histogram - المدرج الإحصائي </center>


```python
sns.distplot(tips['tip'], kde=False)
```

```python
g = sns.FacetGrid(tips, row='sex', col='smoker', size=4)
g.map(plt.hist, 'tip', alpha=0.5, bins=15)
g.add_legend()
plt.show()
```
