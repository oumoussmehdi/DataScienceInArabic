Since seaborn also uses matplotlib to do its plotting you can easily combine the two

and introduces additional plot types. It also makes your traditional Matplotlib plots look a bit prettier.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns

sns.set(style="ticks", 
        color_codes=True)
```

```python
fifa = pd.read_csv('..\\data\\fifa19.csv',index_col=0)
print(fifa.shape)
fifa.head(10)
```

```python
fifa.columns
```

```python
fifa.dtypes
```

### Scattter plot

```python
sns.lmplot(x='Strength', 
           y='Stamina', 
           data=fifa,
          fit_reg=False,
          )
```

### line chart

```python
sns.lineplot(x='Vision', 
             y='Penalties', 
             data=fifa)
```

```python
data =  [1,5,3,2,6] * 20
time = range(100)

sns.tsplot(data, time)
```

### Bar Chart

```python
g = sns.barplot(x="Name", 
            y="Potential", 
            data=fifa[:50],
           )

g.set_xticklabels(rotation=30)
```

```python
g = sns.factorplot("Nationality", 
                   data=fifa, 
                   aspect=1.5, 
                   kind="count", 
                   color="b")
g.set_xticklabels(rotation=90)
```

```python
sns.factorplot(x="Name", 
            y="Potential",
            #hue="class", 
            kind="bar", 
            data=fifa[:20]).set_xticklabels(rotation=90)
```

### Histogram

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
g = sns.FacetGrid(tips, row='sex', col='smoker', size=4)
g.map(plt.hist, 'tip', alpha=0.5, bins=15)
g.add_legend()
plt.show()
```
