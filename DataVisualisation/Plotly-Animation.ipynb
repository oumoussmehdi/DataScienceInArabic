```python
dataset = nations
```

```python
dataset.head()
```

```python
years_from_col = set(dataset['year'])
years_ints = sorted(list(years_from_col))
years = [str(year) for year in years_ints]
#years.remove('1957')
years
```

```python
# make list of regions
regions = []
for region in dataset['region']:
    if region not in regions: 
        regions.append(region)
```

```python
regions
```

```python
columns = []
# make grid
for year in years:
    for region in regions:
        dataset_by_year = dataset[dataset['year'] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[dataset_by_year['region'] == region]
        for col_name in dataset_by_year_and_cont:
            # each column name is unique
            column_name = '{year}_{region}_{header}_gapminder_grid'.format(
                year=year, region=region, header=col_name
            )
            a_column = Column(list(dataset_by_year_and_cont[col_name]), column_name)
            columns.append(a_column)
```
